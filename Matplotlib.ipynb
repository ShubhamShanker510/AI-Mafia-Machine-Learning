## example 7
plotting pie chart

```python
#defining labels
activites=['eat','sleep','work','play']
```

```python
#potion covered by each label
slices=[3,7,8,6]
```

```python
#color for each label
colors=['r','m','g','b']
```

```python
#plotting the pie chat
plt.pie(slices,labels=activites,colors=colors, startangle=90,shadow=True,explode=(0,0,0.1,0),
       radius=1.2,autopct='%1.1f%%')
```

```python
plt.close()
```

## Example 9
plotting subplots using method 1

```python
#function to generate cordinates
def create_plot(ptype):
    ##setting the x-axis value
    x=np.arange(-10,10,0.01)
    
    ##seting the y-xis value
    if ptype=='linear':
        y=x
    elif ptype=='quadratic':
        y=x**2
    elif ptype=='cubic':
        y=x**3
    
    return (x,y)
```

```python
#create a figure
fig=plt.figure()
```

```python
#define subplots ad thier position in figures
plt1=fig.add_subplot(221)
plt2=fig.add_subplot(222)
plt3=fig.add_subplot(223)
plt4=fig.add_subplot(224)
```
